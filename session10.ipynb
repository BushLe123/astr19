```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
random_numbers = np.random.logistic(0, 1, 1000)
```

```python
plt.hist(random_numbers, bins=100)
plt.xlabel("numbers")
plt.ylabel("frequency")
plt.title("histogram of 1000 random numbers between 0-1 from the logistic distribution")
plt.show()
```
